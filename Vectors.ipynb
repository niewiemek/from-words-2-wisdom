```kotlin
import com.google.gson.Gson
import io.ktor.client.*
import io.ktor.client.plugins.logging.*
import io.ktor.client.request.*
import io.ktor.client.statement.*
import io.ktor.http.*
import kotlinx.coroutines.runBlocking

data class EmbeddingRequest(val model: String, val prompt: String)

val client = HttpClient {
    install(Logging) {
        level = LogLevel.NONE
    }
}

suspend fun embedd(text: String) =
    client.post("http://localhost:11434/api/embeddings") {
        contentType(ContentType.Application.Json)
        setBody(
            EmbeddingRequest("mxbai-embed-large", text).let { Gson().toJson(it) }
        )
    }.bodyAsText()

runBlocking {
    val hello = embedd("Hello World")
    hello
}

runBlocking {
    val hello32 = embedd("Hello War")
    hello32
}

```
